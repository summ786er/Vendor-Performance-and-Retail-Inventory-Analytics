```python
vendor_sales_summary=pd.read_sql_query("""
WITH SalesAgg AS (
    SELECT 
        VendorNo, 
        Brand,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(ExciseTax) AS TotalTaxPay
    FROM sales
    GROUP BY VendorNo, Brand
),

InvoiceAgg AS (
    SELECT 
        VendorNumber,
        SUM(Freight) AS TotalFreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber
),

PurchaseAgg AS (
    SELECT
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price AS ActualPrice,
        pp.Volume,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN  purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY 
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price,
        pp.Volume
)

SELECT
    PurchaseAgg.VendorNumber,
    PurchaseAgg.VendorName,
    PurchaseAgg.Brand,
    PurchaseAgg.Description,
    PurchaseAgg.PurchasePrice,
    PurchaseAgg.ActualPrice,
    PurchaseAgg.Volume,
    PurchaseAgg.TotalPurchaseQuantity,
    PurchaseAgg.TotalPurchaseDollars,
    SalesAgg.TotalSalesQuantity,
    SalesAgg.TotalSalesPrice,
    SalesAgg.TotalSalesDollars,
    SalesAgg.TotalTaxPay,
    InvoiceAgg.TotalFreightCost
FROM PurchaseAgg
Inner JOIN SalesAgg
  ON PurchaseAgg.VendorNumber = SalesAgg.VendorNo
 AND PurchaseAgg.Brand = SalesAgg.Brand
Inner JOIN InvoiceAgg
  ON PurchaseAgg.VendorNumber = InvoiceAgg.VendorNumber
ORDER BY TotalSalesDollars DESC
""", conn)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(r"C:\Users\asus\Desktop\Vendor_Analysis\Vendor_DataSales_Analysis\final_inventory.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS vendor_sales_summary (
    VendorNumber INTEGER,
    VendorName VARCHAR(100),
    Brand INTEGER,
    Description VARCHAR(100),
    PurchasePrice DECIMAL(15,2),
    ActualPrice DECIMAL(15,2),
    Volume DECIMAL(15,2),
    TotalPurchaseQuantity INTEGER,
    TotalPurchaseDollars DECIMAL(15,2),
    TotalSalesQuantity DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalSalesDollars DECIMAL(15,2),
    TotalTaxPay DECIMAL(15,2),
    TotalFreightCost DECIMAL(15,2),
    TotalGrossProfit DECIMAL(15,2),
    TotalProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    RemainingQuantity DECIMAL(15,2),
    SalesPurchaseRatio DECIMAL(15,2)
);
""")

conn.commit()

# Insert DataFrame into SQLite
vendor_sales_summary.to_sql(
    name="vendor_sales_summary",
    con=conn,
    if_exists="replace",
    index=False
)

conn.close()

print("Inserted", len(vendor_sales_summary), "filtered rows into Final_Result")
```
